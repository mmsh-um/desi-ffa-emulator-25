### Defining a function to "paint" an healpix map

```python
def paint_healpix_map(NSIDE, RA, dec, w, disc_radius):

    npart = len(RA)
    print('npart =', npart)
    print('NSIDE =', NSIDE)
    print(
    " Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
        )
    )
    NPIX = hp.nside2npix(NSIDE)
    print('NPIX =', NPIX)
    print('disc_radius =', disc_radius)
    
    vector = hp.ang2vec(np.radians(90 - dec), np.radians(RA))
    #print(vec)
    
    hpmap = np.zeros(NPIX)
    pix_count = []
    for n in range(0, npart):
        ipix_disc = hp.query_disc(nside=NSIDE,
                                 vec=vector[n],
                                 radius=np.radians(disc_radius),
                                 nest=True)
        ncenters=len(ipix_disc)
        pix_count.append(ncenters)
        if ncenters > 0:
            hpmap[ipix_disc] = hpmap[ipix_disc] + w[n]/ncenters

    plt.figure(figsize=[4,1])
    plt.hist(pix_count, bins=100);
    plt.title("ncenters")

    return hpmap
```
